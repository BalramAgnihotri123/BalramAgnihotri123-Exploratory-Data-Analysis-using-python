```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_excel("Downloads/tableau files/Zi Cloud Internship Assignment.xlsx")
```

```python
df.head()
```

```python
df['Last Name'] = df['Last Name'].fillna('')
```

```python
campaign_counts = df.groupby("City")['No of Campaign'].sum().reset_index().sort_values("No of Campaign",ascending = False).reset_index()[:15]
```

```python
campaign_counts = df.groupby("City")['No of Campaign'].sum().reset_index().sort_values("No of Campaign",ascending = False).reset_index()[:15]
sns.barplot(data = campaign_counts,
           x = 'City',
           y = 'No of Campaign')
plt.xticks(rotation = 90)
```

* All the major campaigns are done in India that so in major technological hubs like Mumbai, Delhi, Bengaluru etc.

```python
campaign_counts = df.groupby("Industry")['No of Campaign'].sum().reset_index().sort_values("No of Campaign",ascending = False).reset_index()[:15]
sns.barplot(data = campaign_counts,
           x = 'Industry',
           y = 'No of Campaign',
           palette='rainbow')
plt.xticks(rotation = 90)
```

* Most of the camapigns are done by Technological and leisure/ luxary related industries

```python
campaign_counts = df.groupby("Team")['No of Campaign'].sum().reset_index().sort_values("No of Campaign",ascending = False).reset_index()[:15]
sns.barplot(data = campaign_counts,
           x = 'Team',
           y = 'No of Campaign',
           palette='dark')
plt.xticks(rotation = 90)
```

* We can clearly see that HR, Founding, Leadership and Marketing team does most of the campaigns for all the comapanies.
* This is expected from most comapnies as campaigns are generally done by top and middle level of management in an organisation.

* Analyzing code part does not make sense as the code are assigned according to the role of an indiviual i.e, H1 for HR team, P1 for Product team, T1 for technological team etc.
